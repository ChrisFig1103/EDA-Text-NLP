# **Paths**


```python
MY_HOME =  os.getcwd()

CLEANED_NORMALIZED_CORPUS = join(MY_HOME, 'CORPUS', 'cleaned-normalized')

TAGGED_CORPUS = join(MY_HOME, 'CORPUS', 'tagged')
```

# **POS Tagging**


```python
archivos = os.listdir(CLEANED_NORMALIZED_CORPUS)

if archivos:
  for archivo in archivos:
    filename = os.path.join(CLEANED_NORMALIZED_CORPUS, archivo)
    with open(filename, 'r') as of:

      file_content = of.read()
      file_content_tokenizado = word_tokenize(file_content)

      file_content_POS_tagging = nltk.pos_tag(file_content_tokenizado)
      resultado = []
      for tupla in file_content_POS_tagging:
        dummy_string = f"{tupla[0]}/{tupla[1]}"
        resultado.append(dummy_string)

      archivo  = archivo.replace("_cleaned.txt","_tagged.txt")
      filename_tagged = os.path.join(TAGGED_CORPUS, f"{archivo}")

      with open(filename_tagged, 'w') as f:
          f.write(str(resultado))
      print(resultado)
else:
    print("No files found in the directory.")
```
